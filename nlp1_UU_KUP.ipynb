#perubahan pasal

```python
import io
import re
import requests
from PyPDF2 import PdfReader

pdf_url = 'https://drive.google.com/file/d/1DeaDn93lBu0H6RdQvHlmQlf1Ru2PQOOP/view?usp=sharing'
pattern = r'Ketentuan Pasal (\w+).*?diubah'

# Downloading the file from Google Drive
file_id = pdf_url.split('/')[-2]
dwn_url = 'https://drive.google.com/uc?id=' + file_id
response = requests.get(dwn_url)
pdf_data = io.BytesIO(response.content)

# Creating a PDF reader object
reader = PdfReader(pdf_data)

# Extracting text from all pages
text = ''
for page in range(len(reader.pages)):
    text += reader.pages[page].extract_text()

# Searching for the pattern in the text
matches = re.finditer(pattern, text, re.DOTALL)
for match in matches:
    start_index = match.start()
    end_index = match.end()
    pasal_number = match.group(1)
    print(f"Perubahan pada pasal {pasal_number}:")

    # Find the end of the pasal text
    next_pasal_index = text.find("Ketentuan Pasal", end_index)
    if next_pasal_index == -1:
        # No more occurrences of "Ketentuan Pasal" were found, so use the end of the text
        next_pasal_index = len(text)
    pasal_text = text[end_index:next_pasal_index].strip()
    print(pasal_text)
```
